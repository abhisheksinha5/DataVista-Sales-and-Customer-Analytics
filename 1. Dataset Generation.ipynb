```python
import pandas as pd
import random
from faker import Faker
import datetime

# Initialize Faker to generate fake data
fake = Faker()

# Function to generate unique customer IDs
def generate_customer_id():
    return 'C' + ''.join(random.choices('0123456789', k=8))

# Generate customer data
customer_data = {
    "Customer ID": [generate_customer_id() for _ in range(1000000)],  # Adjusted to 100 rows
    "First Name": [fake.first_name() for _ in range(1000000)],  # Adjusted to 100 rows
    "Last Name": [fake.last_name() for _ in range(1000000)],  # Adjusted to 100 rows
    "Age": [random.randint(18, 70) for _ in range(1000000)],  # Adjusted to 100 rows
    "Sex": [random.choice(["Male", "Female"]) for _ in range(1000000)],  # Adjusted to 100 rows
    "Address": [fake.street_address() for _ in range(1000000)],  # Adjusted to 100 rows
    "City": [fake.city() for _ in range(1000000)],  # Adjusted to 100 rows
    "State": [fake.state() for _ in range(1000000)],  # Adjusted to 100 rows
    "Zip Code": [fake.zipcode() for _ in range(1000000)]  # Adjusted to 100 rows
}

# Create DataFrame for customer data
customer_df = pd.DataFrame(customer_data)

# Read product data from Excel file
product_data = pd.read_csv("product_data.csv")

# Generate sales data using customer data and product data
start_date = datetime.datetime(2023, 1, 1)
end_date = datetime.datetime(2023, 12, 31)
sales_data = {
    "Order ID": range(1001, 1001001),  # Adjusted to 100 rows
    "Order Date": [fake.date_time_between(start_date=start_date, end_date=end_date) for _ in range(1000000)],  # Adjusted to 100 rows
    "Product Name": [random.choice(product_data["Product Name"].tolist()) for _ in range(1000000)],  # Adjusted to 100 rows
    "Product Category": [],  # Adjusted to 100 rows
    "Quantity": [random.randint(1, 2) for _ in range(1000000)],  # Adjusted to 100 rows
    "Price per Unit": [random.uniform(500, 5000) for _ in range(1000000)],  # Adjusted to 100 rows
    "Total Cost": [],
    "City": customer_df["City"],  # Use same city data from customer data
    "State": customer_df["State"],  # Use same state data from customer data
    "Zip Code": customer_df["Zip Code"],  # Use same zip code data from customer data
    "Ship Date": [],  # Empty list to store ship dates
    "Delivered Date": [],  # Empty list to store delivered dates
    "Mode of Payment": [random.choice(["Credit Card", "Debit Card", "Cash", "UPI"]) for _ in range(1000000)],  # Adjusted to 100 rows
    "Order Returned (Yes/No)": [random.choice(["Yes", "No"]) for _ in range(1000000)],  # Adjusted to 100 rows
    "Customer Rating (out of 5)": [random.randint(2, 5) for _ in range(1000000)],  # Adjusted to 100 rows
    "Customer ID": customer_df["Customer ID"]  # Use same customer IDs from customer data
}

# Assign product categories based on product names
for product_name in sales_data["Product Name"]:
    category = product_data.loc[product_data["Product Name"] == product_name, "Product Category"].iloc[0]
    sales_data["Product Category"].append(category)

# Calculate Total Cost
for i in range(1000000):  # Adjusted to 100 rows
    total_cost = sales_data["Quantity"][i] * sales_data["Price per Unit"][i]
    sales_data["Total Cost"].append(total_cost)

# Generate ship dates and delivered dates based on order dates
for order_date in sales_data["Order Date"]:
    ship_date = order_date + datetime.timedelta(days=random.randint(1, 2))  # Ship date 1 or 2 days after order date
    delivered_date = ship_date + datetime.timedelta(days=random.randint(4, 7))  # Delivered date 4 to 7 days after ship date
    sales_data["Ship Date"].append(ship_date)
    sales_data["Delivered Date"].append(delivered_date)

# Create DataFrame for sales data
sales_df = pd.DataFrame(sales_data)

# Write sales data and customer data to Excel file
with pd.ExcelWriter("sales_and_customer_data_1million.xlsx") as writer:
    sales_df.to_excel(writer, sheet_name="Sales Data", index=False)
    customer_df.to_excel(writer, sheet_name="Customer Data", index=False)

print("Sales data and customer data saved to sales_and_customer_data_1million.xlsx")
```
